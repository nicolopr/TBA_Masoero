```python
from TBA_masoero import TBA_loop, plot_epsilon_real, plot_epsilon_im, compute_logY0, plot_Y0, save_epsilon, discretize_rapidity
```

# Set up parameters

```python
rapidityMax = 40
rapidityMin = -40

numPoints = 801
r = 0.00001
a = 1

rapidityvals=discretize_rapidity(rapidityMax,rapidityMin, numPoints)[0]
```

# Compute pseudoenergies

```python
e1, e2, e3, e4, e5 = TBA_loop(r,a, rapidityMax,rapidityMin,numPoints)
```

```python
plot_epsilon_real(rapidityvals,e1,e2,e3,e4,e5)
plot_epsilon_im(rapidityvals,e1,e2,e3,e4,e5)
```

```python
Y0=compute_logY0(e1,e2,e3,e4,e5)
plot_Y0(rapidityvals,Y0)
```

# Save data

```python
save_epsilon(r, a, e1, e2, e3, e4, e5)
```

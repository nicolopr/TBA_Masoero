# Set up parameters

```python
def make_vals_r(r_min,r_max,steps):
    rvals=[]
    curr_r=r_min
    while(curr_r<=r_max):
        rvals.append(curr_r)
        curr_r=curr_r*steps
    return rvals

def make_vals_a(a_min,a_max,steps):
    avals=[]
    curr_a=a_min
    while(curr_a<=a_max):
        avals.append(curr_a)
        curr_a=curr_a*steps
    return avals
```

```python
#standard interval for all values of a and r

rapidityMax = 30
rapidityMin = -30
numPoints = 601
rapidityvals=discretize_rapidity(rapidityMax,rapidityMin, numPoints)[0]

#set r
r_min= 0.000000001
r_max = 10
steps_r=100

#set a
a_min=0.001
a_max=1
steps_a=10

r_vals=make_vals_r(r_min,r_max,steps_r)
a_vals=make_vals_a(a_min,a_max,steps_a)
```

# Compute pseudoenergies and save data

```python
for a in a_vals:
    for r in r_vals:
        e1, e2, e3, e4, e5 = TBA_loop(r,a, rapidityMax,rapidityMin,numPoints)
        #try to check if Y0 makes sense...
        Y0=compute_logY0(e1,e2,e3,e4,e5)
        plot_Y0(rapidityvals,Y0)
        save_epsilon(r, a, rapidityvals,e1, e2, e3, e4, e5)
```
